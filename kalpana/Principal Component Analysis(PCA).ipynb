# Implementing Principal Component Analysis

```python
#Reducing dimensions from 200 bands to 120 bands
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 120)
principalComponents = pca.fit_transform(X)
ev = pca.explained_variance_ratio_
```

```python
principalComponents
```

```python
ev
```

```python
#Visualization of Explained variance ratio
```

```python
plt.plot(np.cumsum(ev))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
#Select 66 as the number of components for PCA
```

```python
pca = PCA(n_components = 66)
dt = pca.fit_transform(df.iloc[:, :-1].values)
df1 = pd.concat([pd.DataFrame(data = dt), pd.DataFrame(data = y.ravel())], axis = 1)
df1.columns = [f'PC-{i}' for i in range(1,67)]+['class']
```

```python
df1.head()
```

```python
#Displaying the bands after PCA
```

```python
fig = plt.figure(figsize = (20, 10))
for i in range(1,1+10):
    fig.add_subplot(2,5,i)
    plt.imshow(df1.loc[:, f'PC-{i}'].values.reshape(145,145), cmap='nipy_spectral')
    plt.axis('off')
    plt.title(f'Band - {i}')
plt.savefig('IP_PCA_Bands.png')
```

```python
df1.to_csv('IP_66_PCA.csv', index=False)
```
