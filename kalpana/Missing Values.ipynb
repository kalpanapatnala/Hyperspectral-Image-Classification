#Feature Egineering:
#Feature Egineering concentrate on sepereating input to features And output to label

```python
features = data.iloc[:,0:10].values
```

```python
features
```

#we are seperated input variables and removed output variable

```python
# Here missing values of a perticular column is replaced with the NAN and replaced with their mean and stored it into the imputer 
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values='NaN', strategy='mean',axis=0)
imputer.fit(features[:,[8,9]])
features[:,[8,9]] = imputer.fit_transform(features[:,[8,9]])
```

```python
#copying values to df
df = pd.DataFrame(features)
```

```python
df
```

```python
#Seperating output variable from the dataset
labels = data.iloc[:,-1].values
```
